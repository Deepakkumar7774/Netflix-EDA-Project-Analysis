## 3. ***Data Wrangling***

```python
import pandas as pd

# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Fill missing values for other columns
df['director'].fillna('No Director', inplace=True)
df['cast'].fillna('No Cast', inplace=True)
df['country'].fillna('No Country', inplace=True)
df['date_added'].fillna(pd.Timestamp('1900-01-01'), inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)

# Extract year and month from 'date_added'
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Check if 'duration' column exists before extracting duration_int and duration_type
if 'duration' in df.columns:
    # Extract 'duration_int' and 'duration_type' from 'duration'
    df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)', expand=True)

    # Convert 'duration_int' to numeric
    df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')

    # Fill missing duration values if needed
    df['duration_int'].fillna(0, inplace=True)
    df['duration_type'].fillna('Unknown', inplace=True)

    # Drop the original 'duration' column (if you don't need it anymore)
    df.drop(columns=['duration'], inplace=True)
else:
    # Handle the case where 'duration' column is not present
    # For example, you can create empty columns or skip this step
    df['duration_int'] = 0
    df['duration_type'] = 'Unknown'
    print("Warning: 'duration' column not found in the DataFrame.")


# Calculate the frequency for each genre (split comma-separated values)
genres = df['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)

# Reset index
df.reset_index(drop=True, inplace=True)

# Display the cleaned dataset
df.head()
```

### What all manipulations have you done and insights you found?



1. Converted "date_added" to datetime format.
2. Extracted "year_added" and "month_added" for time-based analysis.
3. Extract 'duration_int' and 'duration_type' from 'duration'
4. Convert 'duration_int' to numeric
5. Fill missing value with proper placeholder for 'director', 'cast', 'country' columns
6. Handle missing value of 'rating' column with the help of mode function

Boxen plots are particularly effective when dealing with datasets that might have long tails or varying levels of distribution, as is often the case with movie durations. In the context of movie ratings, durations can vary widely across different rating categories (e.g., G, PG, PG-13, R), potentially creating significant outliers and complex distribution patterns.
